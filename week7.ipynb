```python
marks = [90,25,67,45,80]

number = 0
for mark in marks:
  number +=1
  if mark >= 60:
    print("%s번 학생은 합격입니다."%number)

  else:
    print("%s번 학생은 불합격입니다."%number)
```

```python
marks = [90,25,67,45,80]

num = 0
for mark in marks:
  print('현재 num : ',num)
  num += 1
  print('업데이트 : ',num)
  print('현재 mark: ',mark)

  if mark > 70 :
    print('pass')
  else:
    print('fail')

  print('***************')

print('반복 완료!')
```
