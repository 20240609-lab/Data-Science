```python
#★★★ 시험 ★★★

def sum_many(*args):
  n_sum = 0
  for i in args:
    n_sum = n_sum + i
  return n_sum


result = sum_many(1,2,3)
print(result)
```

```python
def operations(choice, *args):

  if choice == 'sum':
    result = 0
    for i in args:
      result = result + i
  elif choice == 'mul':
    result = 1
    for i in args:
      result = result * i
  elif choice == 'div':
    result = args[0]
    for i in args[1:]:
      result = result / i
  elif choice == 'sub':
    result = args[0]
    for i in args[1:]:
      result = result - i
  return result

result1 = operations('sum',1,2)
print(result1)
result2 = operations('mul',1,2)
print(result2)
result3 = operations('div',1,2)
print(result3)
result4 = operations('sub',1,2)
print(result4)
```

```python
def answer(choice, *args):
  if choice == '10x':
    result = [10*i for i in args]
  return result

answer('10x',1,2,3)
```

```python
def answer(choice, *args):
  if choice == '10x':
    result = []
    for i in args:
      result.append(i*10)
  return result

answer('10x',1,2,3)
```

```python
def calculator(a, b, cal):
  if cal == 'sum':
    result = a + b
  elif cal == 'sub':
    result = a - b
  elif cal == 'mul':
    result = a * b
  elif cal == 'div':
    if b == 0:
      return "Error"
    result = a / b
  return result

print(f"Sum (5, 3): {calculator(5, 3, 'sum')}")
print(f"Sub (5, 3): {calculator(5, 3, 'sub')}")
print(f"Div (3, 6): {calculator(3, 6, 'div')}")
print(f"Div (3, 0): {calculator(3, 0, 'div')}")
print(f"Mul (5, 3): {calculator(5, 3, 'mul')}")
```

```python
import math

pi = 3.1415
def cylinder_volume(h,r=5):
  volume = pi * r**2 * h
  print('높이는 %d 입니다.' %h)
  print('반지름은 %d 입니다.'%r)

  return volume

print(f"원기둥의 부피는 {cylinder_volume(5,5)}입니다.")
```

```python
#최대값

def max_num(numbers):
  max_value = numbers[0]
  for num in nums:
    if num > max_value:
      max_value = num
  return max_value

nums = [2, 4, 1, 3, 5]
max_result = max_num(nums)
print(f"최대값 : {max_result}")
```

```python
#최소값

def min_num(numbers):
  min_value = numbers[0]
  for num in nums:
    if num < min_value:
      min_value = num
  return min_value


nums = [2, 4, 1, 3, 5]
min_result = min_num(nums)
print(f"최소값 : {min_result}")
```
