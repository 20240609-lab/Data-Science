```python
#Q21. 함수의 초기값을 이용하여 최종 가격을 계산하는 함수를 완성하시오. 5점

#상품 가격 price와 배송비 delivery_fee를 입력받아 최종 결제 금액을 반환하는 get_total_price 함수를 작성하시오. 단, 배송비 delivery_fee의 초기값은 3000으로 설정하시오.

def get_total_price(price, delivery_fee=3000):
  total = price + delivery_fee
  return total

result1 = get_total_price(20000)
result2 = get_total_price(20000,0)

print(result1)
print(result2)
```

```python
#Q22. 숫자 리스트를 입력받아 평균을 반환하는 get_average 함수를 작성하시오. 4.17점

def get_average(numbers):
  avg = sum(numbers)/len(numbers)
  return avg

scores = [70,80,90,100]
result = get_average(scores)

print(result)
```

```python
#Q19. 리스트에서 짝수의 개수를 세는 코드를 완성하시오. 5점

numbers = [3,8,12,15,20,27,30]

count = 0

for num in numbers:
  if num%2==0:
    count += 1

print(count)
```

```python
 #Q18. 점수에 따라 등급을 출력하는 코드를 완성하시오. 5점

#점수 score가 다음 조건에 따라 등급을 출력하도록 코드를 작성하시오.90점 이상: "A"80점 이상 90점 미만: "B"70점 이상 80점 미만: "C"70점 미만: "D"

score = 85

if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
else:
  print("D")
```

```python
#Q20. 리스트 컴프리헨션을 사용하여 60점 이상인 점수만 선택하는 코드를 완성하시오. 3점

scores = [45,67,82,59,90,76]
pass_scores = [score for score in scores if score >= 60]

print(pass_scores)
```
