```python
#문제1

prices = [1500,3200,800,4500,2000]
total = 0

#시험에 ★★★ total = 0
#합을 할땐 0
#곱을 할땐 1

for price in prices:
  total += price

print(f"모든 물건 가격의 총합: {total}")
```

```python
#문제3

orders = [5, 12, 8, 20, 3]

price = 1000

for orders in orders:
  original = orders * price
  if orders >= 10:
    final = original * 0.9
  else:
    final = original
  print(f"총 가격: {final}")
```

```python
#문제4

inventory = {"사과": 10, "포도": 0, "참외": 5, "수박": 0, "딸기": 12}

for fruit, count in inventory.items():
    if count == 0:
        print(f"{fruit}: 재고 없음.")
```

```python
#문제5

numbers = [12, 54, 33, 89, 7, 26, 92, 45]

max_num = numbers[0]

for num in numbers:
  if max_num < num:
    max_num = num

print(f"가장 큰 수: {max_num}")
```
